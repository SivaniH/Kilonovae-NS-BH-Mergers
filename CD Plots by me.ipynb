```python
#data and bin counts
count_gpamg, bins_count_gpamg = np.histogram(g_pam_g, bins=30)
count_fpamg, bins_count_fpamg = np.histogram(f_pam_g, bins=30)

count_gpami, bins_count_gpami = np.histogram(g_pam_i, bins=30)
count_fpami, bins_count_fpami = np.histogram(f_pam_i, bins=30)

count_gpamr, bins_count_gpamr = np.histogram(g_pam_r, bins=30)
count_fpamr, bins_count_fpamr = np.histogram(f_pam_r, bins=30)

count_gpamu, bins_count_gpamu = np.histogram(g_pam_u, bins=30)
count_fpamu, bins_count_fpamu = np.histogram(f_pam_u, bins=30)

count_gpamy, bins_count_gpamy = np.histogram(g_pam_y, bins=30)
count_fpamy, bins_count_fpamy = np.histogram(f_pam_y, bins=30)

count_gpamz, bins_count_gpamz = np.histogram(g_pam_z, bins=30)
count_fpamz, bins_count_fpamz = np.histogram(f_pam_z, bins=30)



# finding the PDF of the histogram using count values
pdf_gpamg = count_gpamg / sum(count_gpamg)
pdf_fpamg = count_fpamg / sum(count_fpamg)

pdf_gpami = count_gpami / sum(count_gpami)
pdf_fpami = count_fpami / sum(count_fpami)

pdf_gpamr = count_gpamr / sum(count_gpamr)
pdf_fpamr = count_fpamr / sum(count_fpamr)

pdf_gpamu = count_gpamu / sum(count_gpamu)
pdf_fpamu = count_fpamu / sum(count_fpamu)

pdf_gpamy = count_gpamy / sum(count_gpamy)
pdf_fpamy = count_fpamy / sum(count_fpamy)

pdf_gpamz = count_gpamz / sum(count_gpamz)
pdf_fpamz = count_fpamz / sum(count_fpamz)



# using numpy np.cumsum to calculate the CDF
cdf_gpamg = np.cumsum(pdf_gpamg)
cdf_fpamg = np.cumsum(pdf_fpamg)

cdf_gpami = np.cumsum(pdf_gpami)
cdf_fpami = np.cumsum(pdf_fpami)

cdf_gpamr = np.cumsum(pdf_gpamr)
cdf_fpamr = np.cumsum(pdf_fpamr)

cdf_gpamu = np.cumsum(pdf_gpamu)
cdf_fpamu = np.cumsum(pdf_fpamu)

cdf_gpamy = np.cumsum(pdf_gpamy)
cdf_fpamy = np.cumsum(pdf_fpamy)

cdf_gpamz = np.cumsum(pdf_gpamz)
cdf_fpamz = np.cumsum(pdf_fpamz)



#plotting
plt.plot(bins_count_gpamg[1:], cdf_gpamg, label="Gaussian Prior", color="blue")
plt.plot(bins_count_fpamg[1:], cdf_fpamg, label="Flat Prior", color="orange")

plt.plot(bins_count_gpami[1:], cdf_gpami, color="blue" )
plt.plot(bins_count_fpami[1:], cdf_fpami, color="orange")

plt.plot(bins_count_gpamr[1:], cdf_gpamr, color="blue" )
plt.plot(bins_count_fpamr[1:], cdf_fpamr, color="orange")

plt.plot(bins_count_gpamu[1:], cdf_gpamu, color="blue" )
plt.plot(bins_count_fpamu[1:], cdf_fpamu, color="orange")

plt.plot(bins_count_gpamy[1:], cdf_gpamy, color="blue" )
plt.plot(bins_count_fpamy[1:], cdf_fpamy, color="orange")

plt.plot(bins_count_gpamz[1:], cdf_gpamz, color="blue" )
plt.plot(bins_count_fpamz[1:], cdf_fpamz, color="orange")

plt.legend()
plt.title("CDF Comparison of Peak Apparent Magnitudes in both BNS Models", fontsize = 10)
```

```python
# No of Data points
N = 1000

# initializing random values
data = np.random.randn(N)

# getting data of the histogram
count, bins_count = np.histogram(data, bins=10)

# finding the PDF of the histogram using count values
pdf = count / sum(count)

# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)

# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```
